```python
20.design a contour plot visualization using matplotlib or seaborn module 
```

```python
# contour plots

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5,5,1000)
y = np.linspace(-5,5,1000)

X,Y = np.meshgrid(x,y)
Z = np.sin((X**2 + Y**2))

plt.contourf(X,Y,Z,cmap='viridis')
plt.colorbar()
```
